```python
# Import Libraries
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
```

```python
# Load the csv data into a dataframe
file_path = ("application_record Adjusted 3 v4.csv")
df = pd.read_csv(file_path)
```

```python
#display sample data
df.head()
```

```python
# Drop Nan
df2 = df.dropna()
```

```python
# We need to use OneHotEncoder for categorical data set
# Create a OneHotEnconder instance
enc = OneHotEncoder(sparse=False)
```

```python
# Create a list of the columns with categorical variables
categorical_variables = ["CODE_GENDER","FLAG_OWN_CAR", "FLAG_OWN_REALTY", "NAME_INCOME_TYPE", "NAME_EDUCATION_TYPE", "NAME_FAMILY_STATUS", "OCCUPATION_TYPE"]
```

```python
# Use the fit_transform function from the OneHotEncoder to encode the data
encoded_data = enc.fit_transform(df2[categorical_variables])
```

```python
# Create a DateFrame with the encoded variables
encoded_df = pd.DataFrame(
    encoded_data,
    columns = enc.get_feature_names(categorical_variables)
)

#Display Date
encoded_df.head()
```

```python
df2.isnull().values.any()
```

```python
# Drop categorical columns from original dataFrame
df2 = df2.drop(columns=["CODE_GENDER","FLAG_OWN_CAR","FLAG_OWN_REALTY","NAME_INCOME_TYPE","NAME_EDUCATION_TYPE","NAME_FAMILY_STATUS","OCCUPATION_TYPE","NAME_HOUSING_TYPE"])

df2.head()
```

```python
# Add the encoded data to the original DataFrame
df2["Gender"] = encoded_df["CODE_GENDER_F"]
df2["Own_Car"] = encoded_df["FLAG_OWN_CAR_N"]
df2["Own_Real_Estate"] = encoded_df["FLAG_OWN_REALTY_N"]
df2["Income_Type"] = encoded_df["NAME_INCOME_TYPE_Commercial associate"]
df2["Education_Type"] = encoded_df["NAME_EDUCATION_TYPE_Academic degree"]
df2["Family_Status"] = encoded_df["NAME_FAMILY_STATUS_Married"]
df2["Occupation_Type"] = encoded_df["OCCUPATION_TYPE_Laborers"]
df2.head()
```

```python
# Drop any nan from new columns
df3 = df2.dropna()
# Check to see if any nan left
df3.isnull().values.any()
```

```python
# Create the features set X
X = df3.drop(columns=["Default"])

# Display sample data
X.head()
```

```python
# Create the target set y
y = df3["Default"]

# Display sample data
y.head()
```

```python
# Split the features and targets set into training and teseting datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a standard scaler instance
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Import counter
from collections import Counter
```

```python
# Under sample the data using "RandomUnderSampler"
from imblearn.under_sampling import RandomUnderSampler
rose = RandomUnderSampler(random_state=1)
X_resampled, y_resampled = rose.fit_resample(X_train, y_train)
Counter(y_resampled)
```

```python
# Fit a Logistic regression model using random undersampled data
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix

y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
# Calculate the Balanced Accuracy Score
from sklearn.metrics import balanced_accuracy_score

balanced_accuracy_score(y_test, y_pred)
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced

print(classification_report_imbalanced(y_test, y_pred))
```

```python
# Fit the data using `ClusterCentroids` and check the count of each class
from imblearn.under_sampling import ClusterCentroids

cc = ClusterCentroids(random_state=1)
X_resampled, y_resampled = cc.fit_resample(X_train, y_train)

from collections import Counter

Counter(y_resampled)
```

```python
# Logistic regression using cluster centroid undersampled data
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver='lbfgs', random_state=1)
model.fit(X_resampled, y_resampled)
```

```python
# Display the confusion matrix
from sklearn.metrics import confusion_matrix
y_pred = model.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
# Calculate the balanced accuracy score
from sklearn.metrics import balanced_accuracy_score

balanced_accuracy_score(y_test, y_pred)
```

```python
# Print the imbalanced classification report
from imblearn.metrics import classification_report_imbalanced
print(classification_report_imbalanced(y_test, y_pred))
```
